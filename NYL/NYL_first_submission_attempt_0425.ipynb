```python
def fit_leastsq_qd(params, data):
    Ddata = (data['Deaths'].values)
    #Idata = (data['TotalCurrentlyPositive'].values)
    s = model_qd(params, data)

    S = s[:,0]
    E = s[:,1]
    I = s[:,2]
    Q = s[:,3]
    R = s[:,4]
    D = s[:,5]
    SA = s[:,6]
    
    error = (D-Ddata)
    return error
```

```python
def seirqd(dat, t, params, N):
    '''
    Modified version with a quarantine factor in dsdt and dedt
    '''
    beta = params[0] / N
    delta = params[1]
    gamma = params[2]
    alpha = params[3]
    lambda_ = params[4]
    kappa = params[5]
    
    s = dat[0]
    e = dat[1]
    i = dat[2]
    q = dat[3]
    r = dat[4]
    d = dat[5]
    sa = dat[6]
    
    Qua = .03
    
    dsdt = - (beta * s * i)*Qua - alpha * s # magical assumption of alpha * s will not be contaged
    dedt = beta * s * i * Qua - gamma * e
    didt = gamma * e - lambda_ * i
    dqdt = lambda_ * i - delta * q - kappa * q
    drdt = delta * q
    dddt = kappa * q
    dsadt = alpha * s
    
    # susceptible, exposed, infected, quarantined, recovered, died, unsusceptible
    return [dsdt, dedt, didt, dqdt, drdt, dddt, dsadt]
```

```python
def model_qd(params, data, tmax=-1):
    # initial conditions --? Where from ??
    N = data['Population'].values[0] # total population
    initial_conditions = N * np.array(params[-5:]) # the parameters are a fraction of the population so multiply by the population
    
    # initial conditions
    e0 = initial_conditions[0]
    i0 = initial_conditions[1]
    q0 = initial_conditions[2]
    r0 = initial_conditions[3]
    sa0 = initial_conditions[4]
    
    d0 = data['Deaths'].values[0]
    s0 = N - np.sum(initial_conditions) - d0

    yz_0 = np.array([s0, e0, i0, q0, r0, d0, sa0]) # dat for seirqd
    
    # Package parameters into a tuple
    args = (params, N)
    
    n = len(data)
    if tmax > 0:
        n = tmax
    
    # Integrate ODEs
    s = scipy.integrate.odeint(seirqd, yz_0, np.arange(0, n), args=args)

    return s
```

```python
# Modified for the temporary LA and NY data


def plot_qd(res, p0_params, p0_initial_conditions, df, region, extrapolate=1, boundary=None, plot_infectious=False):
    data = df
    
    s = model_qd(res.x, data, len(data)*extrapolate) # Integrate and get the predictions for extrapolates
    S = s[:,0]
    E = s[:,1]
    I = s[:,2]
    Q = s[:,3]
    R = s[:,4]
    D = s[:,5]
    SA = s[:,6]

    t = np.arange(0, len(data))
    tp = np.arange(0, len(data)*extrapolate)

    p = bokeh.plotting.figure(plot_width=600,
                              plot_height=400,
                             title =' SEIR-QD Model',
                             x_axis_label = 't (days)',
                             y_axis_label = '# people')

    if plot_infectious:
        p.line(tp, I, color = 'red', line_width = 1, legend_label = 'Model All infected')
        #p.line(tp, S, color = 'purple', line_width = 1.5, legend_label = 'Model Sussceptable')
        p.line(tp, S, color = 'blue', line_width = 1.5, legend_label = 'Model Sussceptable')
        p.line(tp, R, color = 'green', line_width = 1.5, legend_label = 'Model Recover')
        p.line(tp, E, color = 'yellow', line_width = 1.5, legend_label = 'Model Expose')
        p.line(tp, SA, color = 'orange', line_width = 1.5, legend_label = 'Model SA')
    p.line(tp, D, color = 'black', line_width = 1, legend_label = 'Model Dead')

    # death
    p.circle(t, data['Deaths'], color ='black')

    # quarantined
    #p.circle(t, data['TotalCurrentlyPositive'], color ='purple', legend_label='Tested infected')
    
    if boundary is not None:
        vline = Span(location=boundary, dimension='height', line_color='black', line_width=3)
        p.renderers.extend([vline])

    p.legend.location = 'top_left'
    bokeh.io.show(p)
    
    return s
```

```python
def data_merge_preidction(df, s_county, days, date_start = '2020-04-01', date_transition = '2020-04-02'):
    '''
    Return a np array with hybrid counts that has real NYT data before date transition
    and Modeled data of that country after the transition date
    s_county: NP arrary of the prediction
    days: int for how many days from the model arr to be merged into the output
    '''
    actual_arr = df.loc[(df['date']>= date_start) & (df['date']<= date_transition)]['Deaths'].values
    model_use_arr = s_county[:days,5]
    return np.concatenate((actual_arr, model_use_arr), axis= None)
    #return model_use_arr
    
```


def low_death_arr_scaler(df, fips, dummy_day):
    '''
    help scaling the death count to a fake distribution of +- 10 percent when there is low dead count in NYT
    
    '''
    tmp_df = select_county_by_fips(df, fips)
    
    #slope from the available chunk
    s = (tmp_df.iloc[-1]['Deaths'] - tmp_df.iloc[0]['Deaths']) / len(tmp_df)
    death_num = tmp_df.iloc[0]['Deaths'] + s*dummy_day
    
    
    return np.linspace(death_num*.9, death_num*1.1, 9)
    
    

```python
def low_death_predictor(df, fips, days = 91):
    '''
    generate arrays of accumulive death as a linear scaled of two points in the death count
    '''
    tmp_df = select_county_by_fips(df, fips)
    #slope from the available chunk
    
    s = (tmp_df.iloc[-1]['Deaths'] - tmp_df.iloc[0]['Deaths']) / len(tmp_df)
    death_num = tmp_df.iloc[0]['Deaths'] + s*days #death at the end of the prediction
    
    return np.linspace(tmp_df.iloc[0]['Deaths'], death_num, 91)
    
```

def format_output_arr(submission, submission_ids):

```python
def format_output_arr(df, submission, submission_ids):
    
    '''
    Reformat cummulative to non cumulative arr
    function that reads in the submission id, get the fips, look up fips and return
    need to handle high death, low death, and could not find data
    submission: the df that you read csv from the TA's sample
    submission_ids: From the submssion sample format file
    '''
    DAYS = 91
    FIPS_TOT = 3223
    PLACE_HOLDER = np.zeros([FIPS_TOT * 91 , 9]) * 1.0
    
    for day_i in range(DAYS):
        # place holder for one days worth of data
        print()
        print(day_i)
        day_holder = np.zeros([3223, 9]) * 1.0
        
        if day_i == 0: # first entry
            
            for fip_dummy_i in range(FIPS_TOT):
                # iterate thru the submission ids and get the fips for the 
                tmp_fips = correct_FIPS(submission_ids.iloc[fip_dummy_i][11:])
            
                if tmp_fips in low_death_fips:
                    
                    # get the death arr from our dictionary
                    tmp_death_arr = prediction_low_death_dic[tmp_fips]
                    day_holder[fip_dummy_i,:] = high_death_arr_scaler(tmp_death_arr[day_i])

                    # in the high death
                elif tmp_fips in high_death_fips:
                    # get the death from our dictionary
                    tmp_death_arr = prediction_high_death_dic[tmp_fips]
                        # pseudoscale the deaths to +- 10 percent
                    day_holder[fip_dummy_i,:] = high_death_arr_scaler(tmp_death_arr[day_i])
                else:
                        # deal with you later
                    day_holder = day_holder
        
        else: # need to get diff if day > 0
        
            for fip_dummy_i in range(FIPS_TOT):
                # iterate thru the submission ids and get the fips for the 
                tmp_fips = correct_FIPS(submission_ids.iloc[fip_dummy_i][11:])

                # Process the conditionals
                # in the NYT low death
                if tmp_fips in low_death_fips:
                    # deal with you later
                    # get the death from our dictionary
                    tmp_death_arr = prediction_low_death_dic[tmp_fips]
                    day_holder[fip_dummy_i,:] = high_death_arr_scaler(tmp_death_arr[day_i] - tmp_death_arr[day_i-1] )

                # in the high death
                elif tmp_fips in high_death_fips:
                    # get the death from our dictionary
                    tmp_death_arr = prediction_high_death_dic[tmp_fips]
                    # pseudoscale the deaths to +- 10 percent
                    day_holder[fip_dummy_i,:] = high_death_arr_scaler(tmp_death_arr[day_i] - tmp_death_arr[day_i-1])
                else:
                    # deal with you later
                    day_holder = day_holder
                                    
        #Fill into the PLACE HOLDER
        
        PLACE_HOLDER[day_i*FIPS_TOT : (day_i+1) * FIPS_TOT, :] = day_holder
    
    
    #return pd.DataFrame(data = PLACE_HOLDER, index=submission.id.values)
    return pd.DataFrame(data = PLACE_HOLDER, columns = submission.columns[1:], index=submission.id.values)
```
